# 1.4 - Using Column Tranformer and Pipeline

```python

X = data.drop(['class'],axis=1) #,'other_parties','other_payment_plans'

y = data[['class']]

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify=y)

categorical = X_train.dtypes == object

preprocess = make_column_transformer(
        (OneHotEncoder(handle_unknown='ignore'),categorical),
        (StandardScaler(), ~categorical))

select_l = SelectPercentile(score_func=f_regression, percentile=40)
select_k = SelectPercentile(score_func=f_regression, percentile=60)
select_s = SelectPercentile(score_func=f_regression, percentile=50)

skfold = StratifiedKFold(n_splits=5, shuffle=True)

df = y_train
df = df.rename(columns={"class":"status"})
df['status'] = df.status.astype("category").cat.codes
df = df.rename(columns={"status":"class"})
y_train = df


print("Logistic Regression:")
print(cross_val_score(make_pipeline(preprocess,select_l, LogisticRegression(random_state=0)),
                X_train, y_train.values.ravel(), cv=skfold))
print("")

print("KNearestNeighbors Classifier:")
print(cross_val_score(make_pipeline(preprocess,select_k, KNeighborsClassifier()),
                X_train, y_train.values.ravel(), cv=skfold))
print("")

print("Linear SVC:")
print(cross_val_score(make_pipeline(preprocess, select_s, LinearSVC(max_iter=1800)),
                X_train, y_train.values.ravel(), cv=skfold))
```

# 1.5 - Tuning Parameters using GridSearchCV

```python


X = data.drop(['class'],axis=1) #,'other_parties','other_payment_plans'

y = data[['class']]

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify=y)

df = y_train
df = df.rename(columns={"class":"status"})
df['status'] = df.status.astype("category").cat.codes
df = df.rename(columns={"status":"class"})
y_train = df

categorical = X_train.dtypes == object

preprocess = make_column_transformer(
        (OneHotEncoder(handle_unknown='ignore'),categorical),
        (StandardScaler(), ~categorical))


select_k = SelectPercentile(score_func=f_regression, percentile=60)


skfold = StratifiedKFold(n_splits=5, shuffle=True)


knn_pipe = make_pipeline(preprocess,select_k, KNeighborsClassifier())
param_grid = {'kneighborsclassifier__n_neighbors': range(1, 21)}
grid = GridSearchCV(knn_pipe, param_grid, cv=skfold, return_train_score=True)

grid.fit(X_train, y_train.values.ravel())

df = y_test
df = df.rename(columns={"class":"status"})
df['status'] = df.status.astype("category").cat.codes
df = df.rename(columns={"status":"class"})
y_test = df

print('Test Score: ',grid.score(X_test, y_test.values.ravel()))

train_score = grid.cv_results_['mean_train_score']
test_score = grid.cv_results_['mean_test_score']
n_neighbors = grid.cv_results_['param_kneighborsclassifier__n_neighbors']
prev_n_neighbors = n_neighbors

plt.plot(n_neighbors, train_score, 'b', label="mean train scores")
plt.plot(n_neighbors, test_score, 'r', label="mean test scores")

plt.ylabel('Score')
plt.xlabel('n_neighbors')
plt.legend()
plt.title('Scores for varying n_neighbors in KNearestNeighors')
plt.show()
```

```python
X = housing_data.drop(['price','date','street','country'],axis=1)

y = housing_data[['price']]

te = TargetEncoder(cols='statezip')
te.fit(X,y)
X = te.transform(X,y)

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)


categorical = X_train.dtypes == object

cat_preprocessing = make_pipeline(OneHotEncoder(handle_unknown='ignore')) 

cont_preprocessing = make_pipeline(
    SimpleImputer(missing_values=np.nan, strategy='median'),
    StandardScaler())


preprocess = make_column_transformer((cont_preprocessing, ~categorical),(cat_preprocessing,categorical))

kfold = KFold(n_splits=5, shuffle=True)

#########################Ridge#########################################

rid_pipe = make_pipeline(preprocess,Ridge(normalize=True))


param_grid_rid = {'ridge__alpha': np.logspace(-3, 3, 13)}

grid_rid = GridSearchCV(rid_pipe, param_grid_rid, cv=kfold, return_train_score=True)

grid_rid.fit(X_train, y_train)

##########################Lasso##############################

las_pipe = make_pipeline(preprocess, Lasso(normalize=True))

param_grid_las = {'lasso__alpha': np.logspace(-5, 0, 10)}

grid_las = GridSearchCV(las_pipe, param_grid_las, cv=kfold, return_train_score=True)

grid_las.fit(X_train, y_train)

##########################ElasticNet##############################

ela_pipe = make_pipeline(preprocess, ElasticNet())

param_grid_ela = {'elasticnet__alpha': np.logspace(-4, -1, 20)}

grid_ela = GridSearchCV(ela_pipe, param_grid_ela, cv=kfold, return_train_score=True)

grid_ela.fit(X_train, y_train)

#########################Ridge##################################

print('Test Score for Ridge: ',grid_rid.score(X_test, y_test))

train_score_rid = grid_rid.cv_results_['mean_train_score']
test_score_rid = grid_rid.cv_results_['mean_test_score']
alpha_rid = grid_rid.cv_results_['param_ridge__alpha']

#########################Lasso##################################

print('Test Score for Lasso: ',grid_las.score(X_test, y_test))

train_score_las = grid_las.cv_results_['mean_train_score']
test_score_las = grid_las.cv_results_['mean_test_score']
alpha_las = grid_las.cv_results_['param_lasso__alpha']

#########################ElasticNet##################################

print('Test Score for ElasticNet: ',grid_ela.score(X_test, y_test))

train_score_ela = grid_ela.cv_results_['mean_train_score']
test_score_ela = grid_ela.cv_results_['mean_test_score']
alpha_ela = grid_ela.cv_results_['param_elasticnet__alpha']

#########################Ridge##################################

plt.plot(alpha_rid, train_score_rid, 'b', label="mean train score")
plt.plot(alpha_rid, test_score_rid, 'r', label="mean test score")

plt.ylabel('Scores')
plt.xlabel('Param Alpha')
plt.legend()
plt.title('Scores for varying alpha in Ridge')
plt.show()

#########################Lasso##################################

plt.plot(alpha_las, train_score_las, 'b', label="mean train score")
plt.plot(alpha_las, test_score_las, 'r', label="mean test score")

plt.ylabel('Scores')
plt.xlabel('Param Alpha')
plt.legend()
plt.title('Scores for varying alpha in Lasso')
plt.show()

#########################ElasticNet##################################

plt.plot(alpha_ela, train_score_ela, 'b', label="mean train score")
plt.plot(alpha_ela, test_score_ela, 'r', label="mean test score")

plt.ylabel('Scores')
plt.xlabel('Param Alpha')
plt.legend()
plt.title('Scores for varying alpha in ElasticNet')
plt.show()
```
